```python
def extract_song_urls(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    song_urls = []

    # Find all iframe elements
    iframes = soup.find_all('iframe')
    for iframe in iframes:
        src = iframe.get('src', '')
        # Extract the SoundCloud track URL from the src attribute
        url_part = src.split('url=')[1].split('&')[0]
        song_urls.append(f'"{url_part}"')

    # Format the song URLs as a JavaScript array
    js_array = 'const songs = [\n    ' + ',\n    '.join(song_urls) + '\n];'
    return js_array
    
js_array = extract_song_urls(html_content)
print(js_array)
```
